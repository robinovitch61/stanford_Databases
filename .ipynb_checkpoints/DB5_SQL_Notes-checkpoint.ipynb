## Select, Project, & Join

**Introduction:**
* Background terminology (from https://db.grussell.org/section010.html#_Toc67114481)
  * Relation - a set of tuples.
  * Tuple - a collection of attributes which describe some real world entity
    * Informally, an unordered set of linked values and heading-names in which each value has a type and is not Null
  * Attribute - a real world role played by a named domain.
  * Domain - a set of atomic values.
  * Set - a mathematical definition for a collection of objects which contains no duplicates.
  * Schema - set of attribute names
  * Expression - anything that results in a relation
  <br><br>
  
* Example relations used in video: [Relation(columns)]
  * College(cName, state, enrollment)
  * Student(sID, sName, GPA, sizeHS)
  * Apply(sID, cName, major, decision)
  <br><br>

```python
display(Image(filename='initialrelations.png', embed=True, width=800))
```

* Querying other relations results in a relation

* Key for a relation: a set of attributes whose values are guaranteed to be unique

* Simplest query in relational algebra is just the name of a relation
  * The query Student results in a copy of the relation Student
  <br><br>
  
**Select Operator:** used to pick certain rows out of relation(s) $\to$ uses sigma $\sigma_{Condition}Relation$
* $\sigma_{GPA > 3.7}Student$ returns subset of student table containing those rows whose GPA is greater than 7
* $\sigma_{GPA > 3.7\ \wedge\ sizeHS < 1000}Student$ is intersection rows in Student where GPA > 3.7 and sizeHS < 1000
  * Note: $\wedge$ symbol means "and" in a condition (different than intersection $\cap$)
* $\sigma_{cName = 'Stanford'\ \wedge\ major = 'CS'}Apply$ gets applicants to Stanford in CS (rows of Apply)
<br><br>

**Project Operator:** used to pick certain columns out of relation(s) $\to$ uses pi: $\Pi_{Attributes A_1, A_2...A_n}Relation$ where attributes are column names
* $\Pi_{sID, decision}Apply$ gets out two columns (sID and decision) from Apply relation
<br><br>
  
**Combining Extraction of Rows and Columns:**
* Because a query produces a relation, can nest operators:
  * $\Pi_{sID, decision}(\sigma_{GPA > 3.7}Student)$ selects for both columns and rows

**Sets vs. Multisets:**
* SQL: Multisets, bags $\to$ returns duplicates

* Relational Algebra: only returns sets, i.e. no duplicates
<br><br>


**Cross Product or Cartesian Product Operator $\mathbf{X}$:**
* Combines every possible pair of tuples from two relations
  * e.g. if relation A has N tuples and relation B has M tuples, $A X B$ has $N*M$ tuples

```python
display(Image(filename='cartesianprod.png', embed=True, width=500))
```

* Syntactic example: Names and GPAs of students with HS>1000 who applied to CS and were rejected (note: some column names have changed)
$$\Pi_{Name, GPA}(\sigma_{Student.sID = Apply.sID\ \wedge\ HS > 1000\ \wedge\ major = 'CS'\ \wedge\ dec = 'R'}(Student\ X\ Apply))$$
<br>

**Natural Join Operator $\mathbf{\Join}$:**
* Smushes relations together horizontally, enforcing equality on all attributes (columns) with the same name and eliminating duplicate attributes (columns) 

```python
display(Image(filename='naturaljoinex.png', embed=True, width=500))
```

* Not actually doing anything we couldn't just do with cross product and select equality constraints, but convenient notationally

* Example: Names and GPAs of students with HS>1000 who applied to CS and were rejected at college with enr>20000
$$\Pi_{sName, GPA}(\sigma_{HS > 1000\ \wedge\ major = 'CS'\ \wedge\ dec = 'R'\ \wedge\ enr > 20000}(Student \Join Apply \Join College))$$
<br>

**Theta Join Operator $\mathbf{\Join_{\theta}}$:**
* Usually the one database people refer to when they say the "Join" operator

* Specifies the condition for which to keep rows after a cross product:
$$Exp_1 \Join_{\theta} Exp_2 = \sigma_{\theta}(Exp_1\ X\ Exp_2)$$

```python
display(Image(filename='thetajoinex.png', embed=True, width=500))
```

**Union Set Operator $\mathbf{\cup}$:**
* Can combine things "vertically"

```python
display(Image(filename='unionoperatorex.png', embed=True, width=500))
```

* Union Example: List of college and student names:
$$\Pi_{cName}College\ \cup\ \Pi_{sName}Student$$
<br>
* Note that usually schemas (attribute names) are the same for union operator (take care of later)
<br><br>

**Difference Set Operator $\mathbf{-}$:**
* Deletes matching rows

```python
display(Image(filename='differenceoperatorex.png', embed=True, width=500))
```

* Difference example: IDs of students who didn't apply anywhere:
$$\Pi_{sID}Student - \Pi_{sID}Apply$$
<br>
* Can be slightly more complicated, e.g. names and IDs of students who didn't apply anywhere:
$$\Pi_{sID, sName}((\Pi_{sID}Student - \Pi_{sID}Apply) \Join Student)$$
<br><br>

**Intersection Set Operator $\mathbf{\cap}$:**
* Returns intersection where schema and value are equal in two relations

```python
display(Image(filename='intersectionoperatorex.png', embed=True, width=500))
```

* Example: names that are both a college name and student name
$$\Pi_{cName}College\ \cap\ \Pi_{sName}Student$$
  * Note that again, 2 sides of intersection should have same schema (take care of later)
  <br><br>

* Intersection operator equivalent to:
  1. $E_1\ \cap\ E_2 \equiv E_1 - (E_1 - E_2)$
  2. $E_1\ \cap\ E_2 \equiv E_1 \Join E_2 \to$ when $E_1$ and $E_2$ have same schemas
<br><br>

**Rename Operator $\mathbf{\rho}$:**
* Reassign schema names; used to unify schemas for set operators

* Forms:
  1. $\rho_{R(A_1, ...,A_n)}E$ renames relation $E$ to $R$ and renames attributes of $E$ to $A_1, ...,A_n$
  2. $\rho_RE$ just renames $E$ relation to $R$
  3. $\rho_{A_1,....,A_n}E$ renames attributes of $E$ to specified $A_i$s
  <br><br>
  
* Proper syntax for list of college and student names:
$$\rho_{X(name)}(\Pi_{cName}College)\ \cap\ \rho_{X(name)}(\Pi_{sName}Student)$$
<br>

* Also used for disambiguation in "self-joins"
  * e.g. pairs of colleges in same state: $\sigma_{state = state}(College\ X\ College)$ doesn't make explicit sense!
  * Proper way to get pairs of colleges in same state:
  $$\sigma_{s1 = s2}(\rho_{C1(n1, s1, e1)}College\ X\ \rho_{C2(n2, s2, e2)}College)$$
  <br>
  * One more way of getting pairs of colleges in same state:
  $$\sigma_{n1 < n2}(\rho_{C1(n1, s, e1)}College \Join \rho_{C2(n2, s, e2)}College)$$
    * the "<" condition ensures that Berkeley-Berkeley matches are avoided as well as Stanford-Berkeley, Berkeley-Stanford duplicates are avoided
    <br><br>
    
**Assignment Statements:**
* Break down relational algebra statements into its parts

* E.g. pairs of colleges in same state:
\begin{align}
C1 &:= \rho_{C1(n1, s, e1)} \\
C2 &:= \rho_{C2(n2, s, e2)} \\
CP &:= C1 \Join C2 \\
Ans &:= \sigma_{n1 < n2}CP
\end{align}
<br><br>
