## Select, Project, & Join

**Introduction:**
* Background terminology (from https://db.grussell.org/section010.html#_Toc67114481)
  * Relation - a set of tuples.
  * Tuple - a collection of attributes which describe some real world entity
    * Informally, an unordered set of linked values and heading-names in which each value has a type and is not Null
  * Attribute - a real world role played by a named domain.
  * Domain - a set of atomic values.
  * Set - a mathematical definition for a collection of objects which contains no duplicates.
  <br><br>
  
* Example relations used in video: [Relation(columns)]
  * College(cName, state, enrollment)
  * Student(sID, sName, GPA, sizeHS)
  * Apply(sID, cName, major, decision)
  <br><br>
  
* Querying other relations results in a relation

* Key for a relation: a set of attributes whose values are guaranteed to be unique

* Simplest query in relational algebra is just the name of a relation
  * The query Student results in a copy of the relation Student
  <br><br>
  
**Select Operator:** used to pick certain rows out of relation(s) $\to$ uses sigma $\sigma_{Condition}Relation$
* $\sigma_{GPA > 3.7}Student$ returns subset of student table containing those rows whose GPA is greater than 7
* $\sigma_{GPA > 3.7\ \cap\ sizeHS < 1000}Student$ is intersection rows in Student where GPA > 3.7 and sizeHS < 1000
* $\sigma_{cName = 'Stanford'\ \cap\ major = 'CS'}Apply$ gets applicants to Stanford in CS (rows of Apply)
  * Note that from here on, $\cap$ may be written as '^'
<br><br>

**Project Operator:** used to pick certain columns out of relation(s) $\to$ uses pi: $\Pi_{Attributes A_1, A_2...A_n}Relation$ where attributes are column names
* $\Pi_{sID, decision}Apply$ gets out two columns (sID and decision) from Apply relation
<br><br>
  
**Combining Extraction of Rows and Columns:**
* Because a query produces a relation, can nest operators:
  * $\Pi_{sID, decision}(\sigma_{GPA > 3.7}Student)$ selects for both columns and rows

**Sets vs. Multisets:**
* SQL: Multisets, bags $\to$ returns duplicates

* Relational Algebra: only returns sets, i.e. no duplicates
<br><br>


**Cross Product or Cartesian Product Operator $\mathbf{X}$:**
* Combines every possible pair of tuples from two relations
  * e.g. if relation A has N tuples and relation B has M tuples, $A X B$ has $N*M$ tuples

```python
display(Image(filename='cartesianprod.png', embed=True, width=500))
```

* Syntactic example: Names and GPAs of students with HS>1000 who applied to CS and were rejected (note: some column names have changed)
$$\Pi_{Name, GPA}(\sigma_{Student.sID = Apply.sID\ \cap\ HS > 1000\ \cap\ major = 'CS'\ \cap\ dec = 'R'}(Student\ X\ Apply))$$
<br>

**Natural Join Operator $\mathbf{\Join}$:**
* Smushes relations together horizontally, enforcing equality on all attributes (columns) with the same name and eliminating duplicate attributes (columns) 
  * e.g.:

```python
display(Image(filename='naturaljoinex.png', embed=True, width=500))
```

* Not actually doing anything we couldn't just do with cross product and select equality constraints, but convenient notationally

* Example: Names and GPAs of students with HS>1000 who applied to CS and were rejected at college with enr>20000
$$\Pi_{sName, GPA}(\sigma_{HS > 1000\ \cap\ major = 'CS'\ \cap\ dec = 'R'\ \cap\ enr > 20000}(Student \Join Apply \Join College))$$
<br>

**Theta Join Operator $\mathbf{\Join_{\theta}}$:**
* Usually the one database people refer to when they say the "Join" operator

* Specifies the condition for which to keep rows after a cross product:
$$Exp_1 \Join_{\theta} Exp_2 = \sigma_{\theta}(Exp_1\ X\ Exp_2)$$
